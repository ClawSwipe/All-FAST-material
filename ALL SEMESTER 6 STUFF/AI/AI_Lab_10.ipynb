```python

# Importing necessary libraries
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the Iris dataset
data = load_iris()

# Extracting features and target labels
X = data.data
y = data.target

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Visualize the clusters
pca = PCA(n_components=2).fit(X)
X_pca = pca.transform(X)
centers_pca = pca.transform(kmeans.cluster_centers_)

plt.figure(figsize=(8, 6))

# Plot actual species
plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1], c='r', label='Setosa', edgecolors='k')
plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1], c='g', label='Versicolor', edgecolors='k')
plt.scatter(X_pca[y == 2, 0], X_pca[y == 2, 1], c='b', label='Virginica', edgecolors='k')

# Plot clustered data
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='black', marker='x', s=200, label='Cluster Centers')

plt.title('Actual vs Clustered Species')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python

# Importing necessary libraries
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the Wine dataset
data = load_wine()

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(data.data)

# Visualize the reduced-dimensional data
plt.figure(figsize=(8, 6))

for i, target_name in enumerate(data.target_names):
    plt.scatter(X_pca[data.target == i, 0], X_pca[data.target == i, 1], label=target_name)

plt.title('PCA of Wine dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
